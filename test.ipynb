```python
import pandas as pd

df = pd.read_csv('data/tt_data.csv')
df['Experience'] = pd.to_numeric(df['Experience'].str[:-1])*1000
df['Date'] = pd.to_datetime(df['Date'])

df.to_csv('data/tt_data.csv')
```

```python
df['exp_gain']= df.groupby('Name')['Experience'].max()-df.groupby('Name')['Experience'].min()
```
